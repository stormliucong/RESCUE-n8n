```python
from rich import print as rprint
from rich.tree import Tree
from rich.console import Console

def json_to_tree(data, tree):
    if isinstance(data, dict):
        for key, value in data.items():
            branch = tree.add(f"[bold]{key}[/]")
            json_to_tree(value, branch)
    elif isinstance(data, list):
        for i, item in enumerate(data[:3]):  # Limit to first 3 items
            branch = tree.add(f"[cyan][{i}][/]")
            json_to_tree(item, branch)
    else:
        tree.add(repr(data))

tree = Tree("JSON Root")
json_to_tree(response, tree)

console = Console()
console.print(tree)
```

```python
import requests
from datetime import datetime
from typing import Dict, Any

def fetch_and_parse_n8n_execution_log(execution_id: int, webhook_url: str, timeout: int = 10) -> Dict[str, Any]:
    """
    Fetch and parse an n8n execution log into structured evaluation metrics.

    Args:
        execution_id (int): The execution ID to retrieve.
        webhook_url (str): The n8n webhook URL that returns detailed execution logs.
        timeout (int): Timeout for the HTTP request in seconds.

    Returns:
        dict: Extracted information including final output, tiers used, tokens, timings, and tool outputs.
    """
    try:
        resp = requests.get(webhook_url, params={"executionId": execution_id}, timeout=timeout)
        resp.raise_for_status()
        log = resp.json()
    except Exception as exc:
        raise RuntimeError(f"Failed to fetch execution {execution_id}: {exc}")

    run_data = log["data"]["resultData"]["runData"]

    # ----------------------------------------------------------------------
    # 1. Final output extraction (generalized)
    #    Prefer 'Respond to Webhook', else last Agent node that has 'output'
    # ----------------------------------------------------------------------
    final_out = None
    candidate_nodes = ["Respond to Webhook"] + [
        name for name in run_data.keys() if "Agent" in name or "agent" in name
    ]
    for node in candidate_nodes:
        if node in run_data:
            try:
                maybe = run_data[node][0]["data"]["main"][0][0]["json"]
                if "output" in maybe:
                    final_out = maybe["output"]
                    break
            except (KeyError, IndexError, TypeError):
                pass

    # ----------------------------------------------------------------------
    # 2. Dynamic tool tracking (collect all tool-like nodes)
    # ----------------------------------------------------------------------
    tool_outputs = {}
    for node_name, blocks in run_data.items():
        try:
            block = blocks[0]["data"]
            if "ai_tool" in block:
                tool_outputs[node_name] = block["ai_tool"][0][0]["json"]["response"]
            elif "ai_vectorStore" in block:
                tool_outputs[node_name] = block["ai_vectorStore"][0][0]["json"]["response"]
            elif "main" in block:
                possible_json = block["main"][0][0]["json"]
                if isinstance(possible_json, dict):
                    tool_outputs[node_name] = possible_json
        except (KeyError, IndexError, TypeError):
            continue

    tiers_used = list(tool_outputs.keys())

    # ----------------------------------------------------------------------
    # 3. Token tracking across all OpenAI nodes
    # ----------------------------------------------------------------------
    token_total = 0
    for node_name, blocks in run_data.items():
        if node_name.startswith("OpenAI Chat Model"):
            for blk in blocks:
                try:
                    token_total += blk["data"]["ai_languageModel"][0][0]["json"]["tokenUsage"]["totalTokens"]
                except (KeyError, IndexError, TypeError):
                    pass

    # ----------------------------------------------------------------------
    # 4. Input query (best effort fallback)
    # ----------------------------------------------------------------------
    input_query = None
    for node_key in ["Edit Fields1", "Workflow Input Trigger", "Webhook"]:
        if node_key in run_data:
            try:
                input_query = run_data[node_key][0]["data"]["main"][0][0]["json"].get(
                    "chatInput",
                    run_data[node_key][0]["data"]["main"][0][0]["json"].get(
                        "query",
                        run_data[node_key][0]["data"]["main"][0][0]["json"].get(
                            "body", {}
                        ).get("prompt")
                    )
                )
                break
            except (KeyError, IndexError, TypeError):
                pass

    # ----------------------------------------------------------------------
    # 5. Execution timing
    # ----------------------------------------------------------------------
    t0 = datetime.fromisoformat(log["startedAt"].replace("Z", "+00:00"))
    t1 = datetime.fromisoformat(log["stoppedAt"].replace("Z", "+00:00"))
    total_exec_ms = (t1 - t0).total_seconds() * 1000

    # Per-tool execution time
    tool_exec_ms = {}
    for name, blocks in run_data.items():
        try:
            tool_exec_ms[name] = blocks[0]["executionTime"]
        except (KeyError, IndexError, TypeError):
            pass

    # ----------------------------------------------------------------------
    # 6. Tool invocation order (by startTime)
    # ----------------------------------------------------------------------
    tool_order = []
    sortable = []
    for name, blocks in run_data.items():
        try:
            sortable.append((name, blocks[0]["startTime"]))
        except (KeyError, IndexError, TypeError):
            continue
    sortable = sorted(sortable, key=lambda x: x[1])
    tool_order = [name for name, _ in sortable]

    # ----------------------------------------------------------------------
    # 7. Final assembled object
    # ----------------------------------------------------------------------
    return {
        #"raw_log": log,
        "final_out": final_out,
        "tiers_used": tiers_used,
        "token_total": token_total,
        "tool_outputs": tool_outputs,
        "input_query": input_query,
        "total_exec_ms": total_exec_ms,
        "tool_exec_ms": tool_exec_ms,
        "tool_order": tool_order,
    }
```

```python
import requests
from datetime import datetime
from typing import Dict, Any, List, Tuple

# ----------------------------------------------------------------------
# Utility helpers
# ----------------------------------------------------------------------
def _safe_get(d: dict, path: List[str]):
    """
    Walk a list of keys/indices and return value or None.
    """
    cur = d
    for p in path:
        try:
            cur = cur[p]
        except (KeyError, IndexError, TypeError):
            return None
    return cur


# ----------------------------------------------------------------------
# Main parser
# ----------------------------------------------------------------------
def fetch_and_parse_n8n_execution_log(
    execution_id: int,
    webhook_url: str,
    timeout: int = 10,
) -> Dict[str, Any]:
    """
    Generic parser for ANY n8n execution.

    Parameters
    ----------
    execution_id : int
        ID of the execution to fetch.
    webhook_url : str
        The webhook endpoint that returns detailed execution logs when
        called with ?executionId=<id>.
    timeout : int
        HTTP timeout (seconds).

    Returns
    -------
    dict
        {
          "raw_log"        : <full JSON>,
          "final_out"      : str | None,
          "tools_used"     : [node names],
          "token_total"    : int,
          "tool_outputs"   : {node: payload},
          "input_query"    : str | None,
          "total_exec_ms"  : float,
          "tool_exec_ms"   : {node: ms},
          "tool_order"     : [node names]
        }
    """
    # ------------------------------------------------------------------
    # 0. Download execution log
    # ------------------------------------------------------------------
    resp = requests.get(webhook_url, params={"executionId": execution_id}, timeout=timeout)
    resp.raise_for_status()
    log = resp.json()
    run_data = log["data"]["resultData"]["runData"]

    # ------------------------------------------------------------------
    # 1. Calculate chronological node order
    # ------------------------------------------------------------------
    node_times: List[Tuple[str, int]] = []
    for name, blocks in run_data.items():
        start = _safe_get(blocks, [0, "startTime"])
        if isinstance(start, int):
            node_times.append((name, start))
    node_times.sort(key=lambda x: x[1])  # ascending
    tool_order = [n for n, _ in node_times]

    # ------------------------------------------------------------------
    # 2. Determine "final output"
    #     Strategy: iterate nodes in *reverse chronological* order,
    #               grab first JSON with key == "output"
    # ------------------------------------------------------------------
    final_out = None
    for name, _ in reversed(node_times):
        maybe = _safe_get(run_data, [name, 0, "data", "main", 0, 0, "json"])
        if isinstance(maybe, dict) and "output" in maybe:
            final_out = maybe["output"]
            break

    # ------------------------------------------------------------------
    # 3. Gather tool nodes & outputs
    #     We consider a node to be a "tool" if its data block contains
    #     keys like ai_tool, ai_vectorStore, ai_languageModel, or others.
    # ------------------------------------------------------------------
    tools_used, tool_outputs, tool_exec_ms = [], {}, {}
    for name, blocks in run_data.items():
        data_block = _safe_get(blocks, [0, "data"])
        if data_block is None:
            continue

        # capture execution time for every node
        exec_ms = _safe_get(blocks, [0, "executionTime"])
        if isinstance(exec_ms, (int, float)):
            tool_exec_ms[name] = exec_ms

        # identify tool-like payloads
        payload = None
        if "ai_tool" in data_block:
            payload = _safe_get(data_block, ["ai_tool", 0, 0, "json", "response"])
        elif "ai_vectorStore" in data_block:
            payload = _safe_get(data_block, ["ai_vectorStore", 0, 0, "json", "response"])
        elif "ai_languageModel" in data_block:
            payload = _safe_get(data_block, ["ai_languageModel", 0, 0, "json", "response"])
        elif "main" in data_block:  # fallback generic payload
            maybe_json = _safe_get(data_block, ["main", 0, 0, "json"])
            if isinstance(maybe_json, dict):
                payload = maybe_json

        if payload is not None:
            tools_used.append(name)
            tool_outputs[name] = payload

    # ------------------------------------------------------------------
    # 4. Total OpenAI token usage
    # ------------------------------------------------------------------
    token_total = 0
    for name in run_data:
        if name.startswith("OpenAI Chat Model"):
            total = _safe_get(run_data, [name, 0, "data", "ai_languageModel", 0, 0, "json",
                                         "tokenUsage", "totalTokens"])
            if isinstance(total, int):
                token_total += total

    # ------------------------------------------------------------------
    # 5. Best-effort extraction of input query / prompt
    # ------------------------------------------------------------------
    input_query = None
    for node in tool_order:  # search in chronological order
        jq = _safe_get(run_data, [node, 0, "data", "main", 0, 0, "json"])
        if isinstance(jq, dict):
            for key in ("chatInput", "query"):
                if key in jq and isinstance(jq[key], str):
                    input_query = jq[key]
                    break
            if input_query:
                break
            body_prompt = _safe_get(jq, ["body", "prompt"])
            if isinstance(body_prompt, str):
                input_query = body_prompt
                break

    # ------------------------------------------------------------------
    # 6. Total execution wall-clock time
    # ------------------------------------------------------------------
    t0 = datetime.fromisoformat(log["startedAt"].replace("Z", "+00:00"))
    t1 = datetime.fromisoformat(log["stoppedAt"].replace("Z", "+00:00"))
    total_exec_ms = (t1 - t0).total_seconds() * 1000

    # ------------------------------------------------------------------
    # 7. Assemble result
    # ------------------------------------------------------------------
    return {
        "raw_log": log,
        "final_out": final_out,
        "tools_used": tools_used,
        "token_total": token_total,
        "tool_outputs": tool_outputs,
        "input_query": input_query,
        "total_exec_ms": total_exec_ms,
        "tool_exec_ms": tool_exec_ms,
        "tool_order": tool_order,
    }
```
